#### Types of Image Transformations:

1) Affine Transformations - Parallelism between lines is maintained

Example - Scaling, Translation, Rotation

2) Non-Affine Transformations - Parallelism is not maintained, however colinearity between lines is maintained

Example- Projective Transformations, Images from different camera angles

```
import cv2
import numpy as np

from google.colab.patches import cv2_imshow
```

```
image_1 = cv2.imread('image_1.jpg')
cv2_imshow(image_1)
```

```
image_2 = cv2.imread('image_2.jpg')
cv2_imshow(image_2)
```

#### Translation

```
# Translation :

height , width = image_1.shape[:2]
quarter_height , quarter_width = height/4, width/4

print('Shape of Image : ', image_1.shape)

# Translation Matrix T 
T = np.float32(  [ [1, 0, quarter_width], [0, 1, quarter_height]]) 
print('Height : ', height)
print('Width : ', width)

# using warpAffine transform to transform the image using T

img_translated_1 = cv2.warpAffine(image_1, T, (width, height))
cv2_imshow(img_translated_1)
```

#### Rotation

```
# Rotation :

height, width = image_2.shape[:2]

rotation_matrix = cv2.getRotationMatrix2D( (width/2, height/2), 75, 0.7)

rotated_image = cv2.warpAffine(image_2, rotation_matrix, (width, height))

cv2_imshow(rotated_image)
```

```
# now transpose of rotated image
# rotate by 90 degree

rotated_image = cv2.transpose(image_2)
cv2_imshow(rotated_image)
```

```
# Image Pyramids

small_img = cv2.pyrDown(image_1)
large_img = cv2.pyrUp(image_1)

print('\n Image Smaller in dimensions')
cv2_imshow(small_img)
print('\n Image Larger in dimensions')
cv2_imshow(large_img)
```

```
small_img = cv2.pyrDown(image_2)
for i in range(4):
    print('\nSmaller Image')
    cv2_imshow(small_img)
    small_img = cv2.pyrDown(small_img)
```

#### Cropping

```
# Image Cropping using numpy

height, width = image_1.shape[: 2]

start_row, start_column = int(height * 0.25), int(width* 0.25)
end_row, end_column = int(height * 0.75), int(width* 0.75)

cropped_img = image_1[start_row: end_row, start_column: end_column]

print('\nCropped Image')
cv2_imshow(cropped_img)
```

### Arithmetic Operations 

```
# this increases or decreases the brightness from the image

M = np.ones(image_1.shape, dtype = 'uint8') * 100

added = cv2.add(image_1 , M)
subtracted = cv2.subtract(image_1, M)
print("\n Image with Increased Brightness")
cv2_imshow(added)
print("\n Image with Decreased Brightness")
cv2_imshow(subtracted)

```
