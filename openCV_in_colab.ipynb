```
# Split the image

img = cv2.imread('circle.jpg')
B, G, R = cv2.split(img)

print(B.shape)
print(G.shape)
print(R.shape)

print('Actual Image')
cv2_imshow(img)

print("\n B split of Image")
cv2_imshow(B)

print("\n G split of Image")
cv2_imshow(G)

print("\n R split of Image")
cv2_imshow(R)

img_merge = cv2.merge([B, G, R])
print("\n Merged Image")
cv2_imshow(img_merge)
```

```
# Amplify a single channel :

# aplification works such that maximum value is 255 

img_merge_blue = cv2.merge([B + 100, G, R])
print("\n Blue is amplified")
cv2_imshow(img_merge_blue)

img_merge_green = cv2.merge([B , G + 100, R])
print("\n Green is amplified")
cv2_imshow(img_merge_green)

img_merge_red = cv2.merge([B , G, R + 100])
print("\n Red is amplified")
cv2_imshow(img_merge_red)
```

```
# seperate colors as individual images

B, G, R = cv2.split(img)

# create a matrix and add them as two extra dimensions

zeros = np.zeros(img.shape[:2], dtype = 'uint8')

print("\nOnly Red is visible")
cv2_imshow(cv2.merge([zeros, zeros, R]))
print("\nOnly Green is visible")
cv2_imshow(cv2.merge([zeros, G, zeros]))
print("\nOnly Blue is visible")
cv2_imshow(cv2.merge([B, zeros, zeros]))
```
